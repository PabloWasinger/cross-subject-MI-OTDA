```python
import pandas as pd

def show_accuracy_table(csv_path='results/comparison/summary_src_03.csv'):
    """Muestra tabla con métodos en filas y targets en columnas (sin kappa)."""
    df = pd.read_csv(csv_path)
    
    # Filtrar solo columnas de accuracy (sin kappa)
    acc_cols = [c for c in df.columns if not c.endswith('_Kappa') and c != 'Target']
    
    # Transponer: métodos como filas, targets como columnas
    df_acc = df.set_index('Target')[acc_cols].T
    df_acc.columns = [f'S{int(c)}' for c in df_acc.columns]
    
    # Agregar media y std
    df_acc['Mean'] = df_acc.mean(axis=1)
    df_acc['Std'] = df_acc.iloc[:, :-1].std(axis=1)
    
    return df_acc.round(2)

show_accuracy_table()
```

```python
show_accuracy_table(csv_path='results/comparison1/summary_src_03.csv')
```
